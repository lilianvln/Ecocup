```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import random
import os
```

```python
labels = os.listdir(os.path.join("train", "labels_csv"))
```

```python
# Affiche au hasard des images positives de l'ensemble d'apprentissage
fig, axs = plt.subplots(2, 5)
for ax in axs.ravel():
    k = random.randrange(0, len(labels))
    img = plt.imread(os.path.join("train", "images", "pos", labels[k][:-4] + ".jpg"))
    bbox = np.loadtxt(os.path.join("train", "labels_csv", labels[k]), delimiter=",")
    if bbox.ndim == 1:
        bbox = [bbox]

    ax.axis('off')
    ax.set_title(labels[k][:-4])
    ax.imshow(img)
    for bb in bbox:
        rect = patches.Rectangle((bb[1], bb[0]), bb[3], bb[2], linewidth=1, 
                                 edgecolor='r' if bb[4] > 0 else 'g', facecolor='none')
        ax.add_patch(rect)
plt.show()
```
